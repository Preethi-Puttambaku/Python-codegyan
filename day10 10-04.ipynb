```python
#sets (defines with curly brases)
#it is unordered collection 
#by adding wont add at last add by hash value
#set operations 
s = {10,'hi',20,30,'Hi'}
print(s)
```

```python
# set does not allow indexing 
#subscriptable error means that object we caannot do position 
#set methods 
#add 
s1 = {10,2,4,5}
s1.add(7)
print(s1)
#update 
s1.update({20,3})
print(s1)
#remove 
s1.remove(10) #in remove if we try to remove the non exist value gives us the key error
print(s1)
#discard
s1.discard(21) #but in here if we try to remove the non exist it simply prints the set 
print(s1) 
```

```python
s1 = {10,20,30,4}
s2 ={4,5,6,7}
#union
union = s1.union(s2)
print(union)
    #or
print(s1 | s2)
#intersection 
inter=s1.intersection(s2)
    #or 
print(s1 & s2)
print(inter)
#Difference 
print(s1.difference(s2))
#or 
print(s1-s2)
#symmetric difference 
print(s1.symmetric_difference(s2))
#or 
print(s1 ^ s2)
#checking methods 
print(s1.issubset({4,5,6}))
print(s1.issuperset({4,5,6}))
print({1}.isdisjoint(s1))
```
