```python
#string operations
#string concatenation(using + operator)
#used to combine the strings
s1 = "Code"
s2 = "gnan"
s3 = s1 +""+ s2
print(s3)
#string repetition (using * operator)
# to repeat the multiple strings
print(s1*100)
```

```python
#string methods 
#case conversion methods 
#upper '
s = "codegnan is coaching class"
print("upper",s.upper())
#lower
print("lower",s.lower())
#title
print("title",s.title())
#swapcase
print("swap case",s.swapcase())
#capitalize
print(s.capitalize())
```

```python
#if any special characters 
s = "@123s"
print(s.lower())
print(s.upper())
```

```python
#substring index finding 
s = "codegnan"
print(s.find("N"))
print(s.find("n")) #gives the first occurance(n)index value 
print(s.find("gn"))
print(s.index("N"))
```

```python
#white space or pattern removing from string 
#strip()
s = "      COdegnan   "
print(s)
print(s.strip())
#lstrip()
print(s.lstrip())
#rstrip()
print(s.rstrip())
```

```python
s="@@@@#@@ codegnan @# @@"
print(s)
print(s.strip("@"))
print(s.rstrip("@"))
print(s.lstrip("@"))
```

```python
#string Manipulation Methods 
#replace(old,new)
s = "The sky is blue"
print(s.replace("is","isss"))
#split()
print(s.split())
#"sep".join()
lst = ['The', 'sky', 'is', 'blue']
print(" ".join(lst))
```

```python
s ="ABCBCBCA" #It will return when there is no overlapping 
print(s.replace("BC","@"))
print(s.split("A"))
print(s.split("C"))
```

```python
#string checking methods 
#this all methods returns boolean values as result 
#islower()
print("lower","code".islower())
#isupper()
print("upper","CODE".isupper())
#istitle()
print("title","Code".istitle())
print("title","code".istitle())
#isalpha()
print("alpha","@code".isalpha())
print("alpha","code".isalpha())
#isalnum()
print("alnum","code1".isalnum())
#isdigit()
print("digit","123".isdigit())
print("digit","123@".isdigit())
#startswith(substring)
print("startswith","codegnan".startswith("co"))
print("startswith","codegnan".startswith("Co"))
#endswith(substring)
print("endswith","codegnan".endswith("n"))
```

```python
#string comparision 
s1 = "ABC" #it compares with hashkey value 
s2 = "abc"
print(s1 > s2)
print(s1 <s2)
```
