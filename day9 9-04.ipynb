```python
#Tuple
#Immutable(prints in the same order)
t = (1,2,3,4,1,2,"hi",[10,20,30],100)
#indexing 
print(t[7])
print(t)
t[7].append(100) #the length is not changing so we can perform 
#t[0]=10 error because we cannot add items in the tuple 
print(t)
```

```python
#reading tuple of integers 
t = tuple(map(int,input().split()))
print(type(t))
```

```python
#empty tuple
t1 = ()
t2 = tuple()
print(type(t1),type(t2))
```

```python
#Dictionaries 
#empty dictionary representation 
d1 = {}
d3 = dict()
print(type(d1),type(d3))
```

```python
#single value representation 
t1 = (10)
t2 = (5,)
print(type(t1),type(t2))
```

```python
#tuple unpacking
t = (1,2,3)
print(type(t),t)
a,b,c = (10,20,30)
print(a)
```

```python
#tuple packing 
a,*b,c = 10,20,30,40,50 # *b it will store multiple values into it  
print(a,b,c)
```

```python
#dictionary values access 
details = {'name':'sreenu','course':'python','age':20}
print(details['course'])
print(details['Age']) #Age(age) is not found so raises key error 
```

```python
#dictionary keys are unique
details = {'name':'sreenu','course':'python','age':20,'name':"babu"} #it will story only last updated value
print(details)
```

```python
#dictionary is mutable 
details = {'name':'sreenu','course':'python','age':20,'name':"babu"}
details['age'] = 50
details['location'] = 'India' #it will add to the dictonary if the key is not present
print(details)
```

```python
#dictionary methods 
details = {'name':'sreenu','course':'python','age':20,'name':"babu"}
#keys()
keys =details.keys()
values = details.values()
items = details.items()
print(type(keys),keys)
print(type(values),values)
print(type(items),items)
print(list(keys)) 
```

```python
details = {'name':'sreenu','course':'python','age':20,'name':"babu"}
print(details.get('name'))
print(details.get('age',50))
print(details.get('location')) #there is no location but not key error it will print 'none'
print(details.get('location','India')) #here there is no location so returns default value
```

```python
#update
details = {'name':'sreenu','course':'python','age':20,'name':"babu"}
people={'age':50,'education':'10 th','place':"hyderabad"}
print(details.update(people))
print(details)
print(people)
```

```python
#pop key
details = {'name':'sreenu','course':'python','age':20,'name':"babu"}
print(details.pop('name'))
print(details)
#pop item
print(details.popitem())
print(details)
```
